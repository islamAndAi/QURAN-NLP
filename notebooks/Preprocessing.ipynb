```python
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
translation_files = os.listdir("./data/translation")
tafaseer_files = os.listdir("./data/tafaseer")
```

```python
translation_df = []
tafaseer_df = []
translation_author = []
tafaseer_author = []

for i in translation_files:
    translation_df.append(pd.read_csv("./data/translation/" + i))
    translation_author.append(i.split(".")[0].replace("_translation", "").replace("_", " "))
    
for i in tafaseer_files:
    tafaseer_df.append(pd.read_csv("./data/tafaseer/" + i))
    tafaseer_author.append(i.split(".")[0].replace("_tafseer", "").replace("_", " "))
    
```

```python
correct_data_translation = []
correct_translation_author = []
for i in range(len(translation_df)):
    if translation_df[i].shape[0] == 6236:
        correct_data_translation.append(translation_df[i])
        correct_translation_author.append(translation_author[i])
        
correct_data_tafaseer = []
correct_tafaseer_author = []
for i in range(len(tafaseer_df)):
    if not pd.isna(tafaseer_df[i].Arabic.iloc[-1]):
        correct_data_tafaseer.append(tafaseer_df[i])
        correct_tafaseer_author.append(tafaseer_author[i])
```

```python
del translation_df, tafaseer_df
```

```python
Arabic = list(correct_data_tafaseer[0].Arabic)
Name = list(correct_data_translation[0].Name)
Surah = list(correct_data_translation[0].Surah)
Ayat = list(correct_data_translation[0].Ayat)

df = pd.DataFrame(list(zip(Name, Surah, Ayat, Arabic)), columns =['Name', 'Surah', 'Ayat', 'Arabic'])
df
```

```python
for i in range(len(correct_data_translation)):
    df["Translation - " + correct_translation_author[i]] = list(correct_data_translation[i].Verse)

for i in range(len(correct_data_tafaseer)):
    df["Tafaseer - " + correct_tafaseer_author[i]] = list(correct_data_tafaseer[i].Tafseer)
```

```python
surah_df = pd.read_csv("./data/surah/surah_info.csv")
surah_df.head()
```

```python
df = pd.merge(df, surah_df, left_on='Surah', right_on='SurahNumber')
```
