### Fetching data from https://www.thaqalayn-api.net/api/ an API made here https://github.com/MohammedArab1/ThaqalaynAPI

```python
import requests
```

```python
def fetch_books_from_api(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.json()  # Parse JSON response
        else:
            print("Failed to fetch data from API. Status code:", response.status_code)
            return None
    except Exception as e:
        print("An error occurred:", str(e))
        return None

# Example usage
books_data = fetch_books_from_api("https://www.thaqalayn-api.net/api/allBooks")
if books_data:
    print("Number of books fetched:", len(books_data))
    print("First book:", books_data[0])  # Displaying the first book
else:
    print("Failed to fetch books data from the API.")
```

```python
for i in books_data:
    print(i)
```

```python
all_books_data = []

for i in books_data:
    url = "https://www.thaqalayn-api.net/api/" + i['bookId']
    all_books_data.append(fetch_books_from_api(url))
    print("Fetched book data for", i['bookId'], "from the API")
```

## Converting the json data into csv formats

```python
import pandas as pd
```

```python
books_data_df = pd.json_normalize(books_data)
books_data_df.head()
```

```python
books_data_df.to_csv("Thaqalayn_books_data.csv")
```

```python
for i in all_books_data:
    df = pd.json_normalize(i)
    df.to_csv(i[0]["bookId"]+".csv")
    print(i[0]["bookId"], "saved to csv")
```
